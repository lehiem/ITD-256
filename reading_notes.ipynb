### Sock dilemma 

Roommate pairs socks by taking one sock and pulling out many socks from the hamper until he finds a pair. 
The flaw in the system is that it takes ~19 pulls to complete just the first pair so the system is highly inefficient

### Sorting theory 
First and Fundamental insight of sorting theory: scale hurts, sorting a book shelf of 100 is easier than sorting a two book shelves of 50 because you only have to sort once 
- minimizing our pain and suffering when it comes to sorting is all about minimizing the number of things we have to sort
- ex. A way to make sorting socks easier is to have less socks to sort aka do laundry more often

#### Big-O
Big-O notation: Big-O notation provides a way to talk about the kind of relationship that
holds between the size of the problem and the program’s running time. Pg 84
- diving problems into different broad classes

Big-O of one/ O(1) / constant time 
- ex. Hosting means you have to clean but no matter how many people you host the cleaning takes the same amount of time

Big-O of n / O(n)/ linear time 
- n effects the time, higher the n the higher the time 
- time increase by factor of n

Big-O of n-squared/ O(n 2 ) /quadratic time
- n is the power 

Mergesort method 
Split what your sorting into manageable groups, once the smaller groups are sorted join the groups together one by one, if its a group of 6 split into 8 groups than 4 groups of 2 then 2 groups of 2 then merge 

Used by libraries to organize their titles, specifically a large one called the King County Library System

Bucket Sort is what’s used in collaboration with the Mergesort 
- splitting up the books in groups/buckets then organizing within that bucket 

## Chapter 4 - "Caching" up to the heading "The Tyranny of Experience (Pages 113-133)

Space is limited on a computer, you have to manage memory for the sake of time and money

Caching: keeping information that you refer to often in a special place in the memory where it is easily accessible

### Memory hierarchy 
“A hierarchy of memories, each of which has greater capacity than the preceding but which is less quickly accessible.” pg 115

### Library Analogy 
When researching a topic at a library the books you refer to the most are the ones you are going to check out so you have access to them always 
- if you’re able to make just one trip to the stacks to get all the books you need it's almost as good as if every book in the library had already been available at your desk. Pg 116
- The more trips back to the library you make, the slower things go, and the less your desk is really doing for you. Pg 116

### issue with memory
the performance of memory hasn’t gotten much better, which means that relative to processing time, the cost of accessing memory increasing which is not good
- ex. Being able to create something faster in a factory by 2x but still getting the same amount of an necessary import so you can't really make anything 2x more

### Caching Algorithm 
Once a cache fills up bc it can only be a section of the memory what happens?

The goal of a caching algorithm is to limit the amount of times you cant find what you are looking for in the cache so when the cache is full you delete the information that you will not need as soon
Random Eviction: adding new data to the cache and getting rid of old data randomly 

First In, First out (FIFO): evict the first thing added when space is need

Least Recently Used (LRU): taking out the information that hasn’t been accessed to make space for the new information *this is what is mainly used

### Library reference 
Libraries use LRU to decide what books get displayed and which books are held at an offsite/non display area
- if a item isn't used in 12 years UC Berkeley's library sends it off site

Some libraries show their most recently acquired books which is an interaction of FIFO (first in first out)

### issues with caching 
Obviously LRU makes sense but its not applicable all the time, you don't want to throw out an old sentimental shirt even if you don't wear it anymore 

Trying to cache something near the information it applies to like having running gear near the front door for easy access 

Multi-level memory leads to caching on different levels 
- ex. closet being the first level/most accessible, then a basement, than a storage locker
-- in decreasing order to apply LRU
